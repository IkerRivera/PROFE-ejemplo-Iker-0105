```python
#### clasw automovil , atributos, marca, modelo, placa, precio
#  funciones leer datos (), mostrar datos()
# crear 3 objetos carro1, carro2, carro3

class Automovil:
    def __init__(self, marca, modelo, placa, precio):
        self.marca = marca
        self.modelo = modelo
        self.placa = placa
        self.precio = precio

    def leer_datos(self):
        self.marca = input("Ingrese la marca del automóvil: ")
        self.modelo = input("Ingrese el modelo del automóvil: ")
        self.placa = input("Ingrese la placa del automóvil: ")
        self.precio = float(input("Ingrese el precio del automóvil: "))

    def mostrar_datos(self):
        print("Marca:", self.marca)
        print("Modelo:", self.modelo)
        print("Placa:", self.placa)
        print("Precio:", self.precio)

# fin de la classe automovil

# creando los objetos  carro1, carro2, carro3
print("carro num1")
carro1 = Automovil("", "", "", 0.0) # Initialize with default values
carro1.leer_datos()
carro1.mostrar_datos()
print("-*-*-*-*-*-*-*-*-*-*-*-*-*-*")

print("carro num2")
carro2 = Automovil("", "", "", 0.0) # Initialize with default values
carro2.leer_datos()
carro2.mostrar_datos()
print("-*-*-*-*-*-*-*-*-*-*-*-*-*-*")

print("carro num3")
carro3 = Automovil("", "", "", 0.0) # Initialize with default values
carro3.leer_datos()
carro3.mostrar_datos()
print("-*-*-*-*-*-*-*-*-*-*-*-*-*-*")

print("programa realizado por Iker Montoya NL = 38")
```
